```python
#Si el empleado es mayor de 55 años disfrutará de un bono de prepensión correspondiente al 5% de su sueldo básico.
 
edad = int(input("Ingrese la edad del empleado: "))
sueldo = int(input("Ingrese el sueldo del empleado: "))
if edad > 55:
  bono = sueldo *  0.05
  print("El empleado tiene derecho a un bono de prepensión de", bono)
else:
  print("El empleado no tiene derecho a bono de prepensión")
```

```python
#Si el empleado es casado y tiene hijos se le otorgará un paseo cada diciembre


estado_civil = input("¿Está casado? (sí o no): ").lower() == "si"
tiene_hijos = input("¿Tiene hijos? (sí o no): ").lower() == "si"

if estado_civil  and tiene_hijos:
  print("Se otorgará un paseo en diciembre")
else:
  print("No es elegible para el paseo en diciembre")


```

```python

#Si el sueldo básico está entre 1000000 y 1500000 tendrá una 
#comisión del 2% sobre el valor del sueldo; Si el sueldo básico 
#está entre 1500001 y 2000000 tendrá una comisión del 5% sobre 
#el valor del sueldo; para todos los demás casos no habrá comisión.

sueldobas = float(input("Ingrese el sueldo básico del empleado: "))
comision = 0.0

if sueldobas >= 1000000 and sueldobas<= 1500000:
  comision = 0.02
elif sueldobas > 1500000 and sueldobas <= 2000000:
  comision = 0.05

if comision > 0:
  valor_comision = sueldobas * comision
  print("La comisión correspondiente es de", valor_comision)
else:
  print("No hay comisión")
```
