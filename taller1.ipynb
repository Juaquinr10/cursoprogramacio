```python
# Un vendedor recibe un sueldo base más un 10% extra por
# comisión de sus ventas, el vendedor desea saber cuánto
# dinero obtendrá por concepto de comisiones por las tres
# ventas que realiza en el mes y el total que recibirá en
# el mes tomando en cuenta su sueldo base y comisiones.


sueldo = float(input("ingresa sueldo basico "))
venta1 = float(input("ingresa venta 1 : "))
venta2= float(input("ingresa venta 2: "))
venta3 = float(input("ingresa venta 3: "))
comi = ((venta1+venta2+venta3)* 0.10)
print(f"la comision de las 3 ventas es de = {comi}")
total = (sueldo+ comi)
print(f"sueldo total y comision es = {total}")
```

```python
#Un alumno desea saber cuál será su calificación final en
#la materia de Algoritmos. Dicha calificación se compone
#de tres exámenes parciales.


nota1 = float(input("Digite la calificación 1: "))
nota2 = float(input("Digite la calificación 2: "))
nota3 = float(input("Digite la calificación 3: "))


calificacion = round((nota1 + nota2 + nota3) / 3 , 2)

print(f"su calificacion final es de = {calificacion}")
```

```python
# Una tienda ofrece un descuento del 15% sobre el total
# de la compra y un cliente desea saber cuánto deberá
# pagar finalmente por su compra.


valor = float(input("ingresa el total de la compra "))
descuento = valor * 0.15
total = valor - descuento

print(f"el descuento aplicado a la compra es = {descuento}")
print(f"el total de su copra es = {total}")
```

```python
# Un maestro desea saber qué porcentaje de hombres y que
#porcentaje de mujeres hay en un grupo de estudiantes.

mujeres = float(input("ingresa el total de mujeres "))
hombres = float(input("ingresa el total de hombres "))

totalM = mujeres / (mujeres+ hombres) * 100 ,"%"
totalH = hombres / (mujeres+ hombres) * 100 ,"%"

print(f"el total de mujeres en el salon es de = {totalM}")
print(f"el total de hombres en el salon es de = {totalH}")
```
